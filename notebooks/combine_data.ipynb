```python
import pandas as pd
import numpy as np
import glob 
import re
import pickle
```

```python
out_prefix = '../../oligos/results/oligo_1/counts_combined/oligo_1_10mers_combined'
data_dir = '../../oligos/results/oligo_1/counts/'
k = 10
```

```python
files = glob.glob(f'{data_dir}/*{k}mer*')
files
```

```python
len(files)
```

```python
df = pd.read_csv(files[0])
df
```

```python
for i in files:
    name = re.search('([^\/]+$)', i)[0]
    meta_data = name.split('_')
    print(meta_data)
```

```python
df_list = []
for i in files:
    # getting meta data
    name = re.search('([^\/]+$)', i)[0]
    meta_data = name.split('_')
    substrate = '_'.join([meta_data[0], meta_data[1]])
    condition = meta_data[2]
    rep = meta_data[3].replace('rep','')
    k = meta_data[4].replace('mers','')
    # add meta data to the table
    df = pd.read_csv(i)
    df[['substrate','condition','rep','k']] = [substrate, condition, rep, k]
    df_list.append(df)
```

```python
df_list[1]
```

```python
combined_long = pd.concat(df_list).reset_index(drop=True)
# convert nodes to edge
combined_long['edge'] = combined_long['node1']+combined_long['node2'].str.rstrip().str[-1]
combined_long = combined_long.drop(['node1','node2'], axis=1)
```

```python
combined_long
```

```python
# pivot count data to long format
combined_wide = combined_long.pivot(index=['edge'], columns=['condition','rep'], values='count').reset_index()
```

```python
combined_wide
```

```python
# flatten multi-index columns
combined_wide.columns = combined_wide.columns.map('_'.join).str.strip('_')
# reorder columns lexicographically
combined_wide = combined_wide.reindex(sorted(combined_wide.columns), axis=1)
# move node columns back to front
combined_wide = combined_wide.set_index(['edge']).reset_index()
# convert floats back to int
num_cols = combined_wide.columns[combined_wide.dtypes.eq('float64')]
combined_wide[num_cols] = combined_wide[num_cols].fillna(0).astype(int)
```

```python
combined_wide
```

```python
combined_wide.to_csv(f'{out_prefix}_wide.csv', index=False)
```
