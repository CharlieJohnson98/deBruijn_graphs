```python
import pandas as pd
import glob
import os
```

```python
enr_file = '../oligo/results/10mer/enrichment_all/oligo_3_3w_enrich.csv'
```

```python
df = pd.read_csv(enr_file)
df
```

```python
for k in range(14, 16):
    files = glob.glob(f'../oligo/results/{k}mer/enrichment_all/*enrich.csv')
    for f in files:
        print(f)
        # read in data
        df = pd.read_csv(f)
        # filter
        df_z = df[df.apex_zscore >= 1.96]
        df_fdr = df[df.apex_pval_bh <= 0.05]
        # write out results
        f_name = os.path.basename(f)
        f_name = f_name.replace(".csv","")
        f_out_z = f'../oligo/results/{k}mer/enrichment_all/{f_name}_z_filter.csv'
        f_out_fdr = f'../oligo/results/{k}mer/enrichment_all/{f_name}_fdr_filter.csv'
        df_z.to_csv(f_out_z, index=False)
        df_fdr.to_csv(f_out_fdr, index=False)
```
